```python
import pandas as pd
import os

# Step 1: Get current directory
BASE_DIR = os.getcwd()

# Step 2: Move to project root (go up from notebooks)
BASE_DIR = os.path.dirname(BASE_DIR)

print("Fixed Base Dir:", BASE_DIR)

# Step 3: Create correct paths
fake_path = os.path.join(BASE_DIR, "data", "raw", "Fake.csv")
true_path = os.path.join(BASE_DIR, "data", "raw", "True.csv")

print("Fake Path:", fake_path)
print("Exists:", os.path.exists(fake_path))

# Step 4: Load datasets
fake_df = pd.read_csv(fake_path)
true_df = pd.read_csv(true_path)

# Step 5: Labels
fake_df["label"] = 0
true_df["label"] = 1

# Step 6: Merge + Shuffle
df = pd.concat([fake_df, true_df], axis=0)
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Step 7: Info
print(df.info())
print(df["label"].value_counts())
print(df.head())
```

```python
import re
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))

def clean_text(text):
    # Lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Tokenization
    tokens = word_tokenize(text)
    
    # Remove stopwords
    tokens = [word for word in tokens if word not in stop_words]
    
    return " ".join(tokens)
```

```python
df["clean_text"] = df["text"].apply(clean_text)
```
